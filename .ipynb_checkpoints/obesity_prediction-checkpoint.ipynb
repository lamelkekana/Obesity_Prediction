<div style="text-align: center; border-top: 3px solid black; padding: 10px;">
    <h2><strong>4.Data Cleaning</strong></h2>
</div>

After using the info() and describe() functions for initial data inspection, we saw that there were no missing values. However, we will verify this further by explicitly checking for any missing values. Additionally, we will check for any duplicate rows in the dataset to ensure the data is clean.

```python
# verify if theres no missing rows

no_missing_rows = df.isna().sum()

print(no_missing_rows)
```

```python
# Number of duplicate rows
duplicate_count_before = df.duplicated().sum()

# print duplicate count
print('Duplicate row count :',duplicate_count_before)

# remove duplicate rows
df = df.drop_duplicates()

# Number of duplicate rows after deleting
duplicate_count_after = df.duplicated().sum()

# print duplicate count
print('Duplicate row count after dropping duplicates:',duplicate_count_after)
```

    Besides the dataset having duplicates, no missing values were found. After removing the duplicates, the dataset is now clean and free of any missing data and it is now ready for further analysis

    <h3><strong>4.1 Univariate Analysis</strong></h3>

```python
# Univariate analysis - Categorical features (display all categorical features with 5 per row)
categorical_columns = ['Gender', 'family_history', 'FAVC', 'CAEC', 'MTRANS', 'Obesity']

# Set up subplots: 1 row, 6 columns
fig, axes = plt.subplots(1, 6, figsize=(18, 5))  # Wide figure to fit all plots

glossy_style = {'alpha': 0.8, 'edgecolor': 'black'} 

for i, col in enumerate(categorical_columns):
    sns.countplot(x=df[col], ax=axes[i], color='skyblue', **glossy_style)
    axes[i].set_title(f'{col}')
    axes[i].tick_params(axis='x', rotation=90)

plt.tight_layout(pad=1.5)  # Adjust padding for better spacing
plt.show()
```

- __Gender:__ The dataset seems to have a balanced distribution between Male and Female respondents.

- __Family History:__ A significant majority of individuals have a family history of obesity, which could be an important predictive factor.

- __Frequent Consumption of High-Calorie Food (FAVC):__ Most individuals in the dataset consume high-calorie food frequently, which aligns with obesity risk factors.

- __Consumption of Food Between Meals (CAEC):__ The majority of people in the dataset consume food "Sometimes," with smaller groups reporting frequent or always snacking.

- __Mode of Transportation (MTRANS):__ Most individuals use public transportation or walk, with fewer using motorbikes, automobiles, or bicycles.

- __Obesity:__ The distribution of obesity levels appears to be relatively balanced, though some categories have slightly higher counts,this is good as this featur will be later used as our target feature

```python
# Univariate analysis - Numerical features (display all numerical features with 4 per row)
numerical_columns = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']
fig, axes = plt.subplots(2, 4, figsize=(16, 8))  # 2 rows, 4 columns

for i, col in enumerate(numerical_columns):
    ax = axes[i // 4, i % 4]  # Calculate row and column
    sns.histplot(df[col], kde=True, ax=ax,color='skyblue')
    ax.set_title(f'Distribution of {col}')

plt.tight_layout(pad=1.0)  # Adjust padding between plots
plt.show()
```

-  __Age:__ Presents a right-skewed plot meaning most individuals are relatively young,most being less than 30 years
- __NCP (Number of Main Meals per Day):__ Strong peaks at 3 meals per day, likely the standard.
- __FAF (Physical Activity Frequency per Week):__ Right-skewed, meaning many individuals have low physical activity.
- __FCVC (Frequency of Vegetable Consumption):__ Suprisingly Many individuals have high vegetable consumption.

```python
# Calculate correlation matrix for numerical columns
orr_matrix = df[numerical_columns].corr()


plt.figure(figsize=(8, 6))  # Adjust figure size

# Create and display the heatmap with annotations and color map
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap") 
plt.show()
```

- __Height and Weight:__ A moderate positive correlation (0.46) is expected. Taller people tend to weigh more.
- __Age and TUE (Time Using Electronic devices):__ A weak negative correlation (-0.30) might suggests that older individuals tend to spend less time using electronic devices.

While many of the numerical features exhibit weak correlations with each other, possibly due to non-linear relationships or their inherent independence, we will explore potential patterns by analyzing them alongside the categorical features.

```python
# Set figure size
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Define blue spectrum colors
colors = ['skyblue', 'dodgerblue', 'deepskyblue']

glossy_style = {'alpha': 0.8, 'edgecolor': 'black'}  # Glossy effect

# Plot 1: Physical Activity vs Obesity Level
sns.barplot(ax=axes[0], x='Obesity', y='FAF', data=df, color=colors[0], **glossy_style)
axes[0].set_title('Physical Activity vs Obesity Level')
axes[0].tick_params(axis='x', rotation=90)

# Plot 2: Means of Transportation vs Weight
sns.barplot(ax=axes[1], x='MTRANS', y='Weight', data=df, color=colors[1], **glossy_style)
axes[1].set_title('Means of Transportation vs Weight')
axes[1].tick_params(axis='x', rotation=90)

# Plot 3: Calorie Consumption vs Weight
sns.barplot(ax=axes[2], x='FAVC', y='Weight', data=df, color=colors[2], **glossy_style)
axes[2].set_title('Calorie Consumption vs Weight')
axes[2].tick_params(axis='x', rotation=90)

# Adjust layout for better spacing
plt.tight_layout()

# Display the plots
plt.show()
```

##### **Physical Activity vs Obesity Level**  
- Normal weight individuals tend to have the highest **physical activity frequency (FAF)**.  
- Obesity Type III individuals show the lowest **Physical Activity**, indicating a strong inverse relationship between physical activity and obesity.  
- Overweight and Obesity Type I & II groups have moderate **Physical Activity**, suggesting a declining trend as weight increases.  

##### **Means of Transportation vs Weight**  
- **Public transportation users** appear to have a **higher weight on average** compared to those who walk.  
- **Walking and biking** are associated with **lower weights**, suggesting that active transportation helps manage weight.  
- **Automobile and motorbike users** fall somewhere in the middle, but still tend to have **higher weights than walkers**.  

##### **Calorie Consumption vs Weight (FAVC - Frequent Consumption of High-Calorie Foods)**  
- Individuals who **frequently consume high-calorie foods ("yes") have a significantly higher weight** than those who do not.  
- This clearly shows the impact of **excessive calorie intake on weight gain**, aligning with obesity risk factors.  
 
âœ… **Physical activity inversely correlates with obesity levels.**  
âœ… **Active transport methods like walking and biking are linked to lower weight.**  
âœ… **Frequent high-calorie food consumption contributes to weight gain.**  


```python
# Define height categories based on observed values
def categorize_height(height):
    if height < 1.60:
        return 'Short'
    elif 1.60 <= height < 1.75:
        return 'Medium'
    else:
        return 'Tall'

# Apply the function to create a new column
df['Height_Category'] = df['Height'].apply(categorize_height)
```

```python
# Define age categories
def categorize_age(age):
    if age < 25:
        return 'Young'
    elif 25 <= age < 45:
        return 'Adult'
    elif 45 <= age <= 60:
        return 'Mid-age'
    else:
        return 'Old'

# Apply the function to create a new column
df['Age_Category'] = df['Age'].apply(categorize_age)
```

```python
# Calculate the average weight by Age_Category and Height_Category
avg_weight_age = df.groupby('Age_Category')['Weight'].mean()
avg_weight_height = df.groupby('Height_Category')['Weight'].mean()

# Create a color palette using blues
colors_age = sns.light_palette("royalblue", n_colors=len(avg_weight_age), reverse=True)
colors_height = sns.light_palette("deepskyblue", n_colors=len(avg_weight_height), reverse=True)

# Create the figure and axes for two pie charts
fig, axes = plt.subplots(1, 2, figsize=(12, 8))

# Plot Average Weight by Age Category on the first axis
axes[0].pie(avg_weight_age, labels=avg_weight_age.index, autopct='%1.0f%%', startangle=90, colors=colors_age)
axes[0].set_title('Average Weight Distribution by Age Category', y=0.8)
axes[0].axis('equal')  # Equal aspect ratio ensures the pie chart is drawn as a circle

# Plot Average Weight by Height Category on the second axis
axes[1].pie(avg_weight_height, labels=avg_weight_height.index, autopct='%1.0f%%', startangle=90, colors=colors_height)
axes[1].set_title('Average Weight Distribution by Height Category', y=0.8)
axes[1].axis('equal')

# Adjust layout to prevent overlap
plt.subplots_adjust(wspace=1)

# Show the plot
plt.show()
```

#### __Average Weight Distribution by Age Category__
- Most age categories show similar average weights, except for the older age group, which stands out with a significantly lower average weight.
- The Old category has the lowest average weight, which could be due to various factors such as aging, changes in metabolism, and health conditions

#### __Average Weight by Height Category__
- Taller people generally have larger frames and more muscle mass, which contributes to higher body weight. 
- Shorter individuals might have smaller frames and less muscle mass on average, leading to lower weight.

```python

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Label encode the 'Obesity' 
df['Obesity'] = label_encoder.fit_transform(df['Obesity'])

# Label encoding for binary columns
df['Gender'] = label_encoder.fit_transform(df['Gender'])
df['family_history'] = label_encoder.fit_transform(df['family_history'])
df['SMOKE'] = label_encoder.fit_transform(df['SMOKE'])
df['FAVC'] = label_encoder.fit_transform(df['FAVC'])
df['SCC'] = label_encoder.fit_transform(df['SCC'])

# One-hot encoding for multi-class columns, returning 0/1 instead of True/False
df_encoded = pd.get_dummies(df, columns=['MTRANS', 'CALC', 'CAEC'], drop_first=True)

# Convert only the one-hot encoded columns to integers
df_encoded[df_encoded.columns[df_encoded.columns.str.startswith(('MTRANS', 'CALC', 'CAEC'))]] = df_encoded[df_encoded.columns[df_encoded.columns.str.startswith(('MTRANS', 'CALC', 'CAEC'))]].astype(int)

df_encoded
```
