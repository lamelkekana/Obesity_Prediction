</div>

The target variable will be encoded using `LabelEncoder`, while features will be encoded based on the number of classes. Features with only two classes will be encoded using `LabelEncoder`, and features with multiple classes will be encoded using `One-Hot Encoding`. This approach is used because `LabelEncoder` is more suitable for binary classification, while `One-Hot Encoding` avoids creating `ordinal` relationships for multi-class features.

```python

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Label encode the 'Obesity' 
df['Obesity'] = label_encoder.fit_transform(df['Obesity'])

# Get the mapping of labels to numbers
label_to_number = dict(zip(label_encoder.classes_, range(len(label_encoder.classes_))))

print("Label to Number Mapping:", label_to_number)


# Label encoding for binary columns
df['Gender'] = label_encoder.fit_transform(df['Gender'])
df['family_history'] = label_encoder.fit_transform(df['family_history'])
df['SMOKE'] = label_encoder.fit_transform(df['SMOKE'])
df['FAVC'] = label_encoder.fit_transform(df['FAVC'])
df['SCC'] = label_encoder.fit_transform(df['SCC'])

# One-hot encoding for multi-class columns, returning 0/1 instead of True/False
df_encoded = pd.get_dummies(df, columns=['MTRANS', 'CALC', 'CAEC'], drop_first=True)

# Convert only the one-hot encoded columns to integers
df_encoded[df_encoded.columns[df_encoded.columns.str.startswith(('MTRANS', 'CALC', 'CAEC'))]] = df_encoded[df_encoded.columns[df_encoded.columns.str.startswith(('MTRANS', 'CALC', 'CAEC'))]].astype(int)
```

```python
# Reverse the dictionary to map numbers back to labels
number_to_label = {v: k for k, v in label_to_number.items()}

# Use number_to_label to get target names
target_names = [number_to_label[i] for i in range(len(label_to_number))]
```

    The dimensionality of the dataset has increased from 17 to 24 columns, with 7 additional columns added.

```python
# Initialize features(X) and target variable(y)
X = df_encoded.drop('Obesity',axis=1) # Features
y = df_encoded['Obesity'] # Target Variable

# Split data in train and test set,with 75% train and 25% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

Only continuous features will be scaled, while categorical features (One-Hot encoded) will be left as is because scaling is typically applied to numerical data to normalize the range and distribution, ensuring that continuous variables have a mean of 0 and standard deviation of 1. Categorical variables, after One-Hot encoding, are already in binary form and do not require scaling, as they represent discrete categories rather than continuous quantities

```python
# Initialize Scaler
scaler = StandardScaler()

# features to be scale
continous_features = ['Age','Height','Weight','FCVC','NCP','CH2O','FAF','TUE']

# Step 1: Scale only the continuous features in X_train
X_train_scaled = X_train.copy()  # Copy to keep original intact
X_train_scaled[continous_features] = scaler.fit_transform(X_train[continous_features])

# Step 2: Scale the continuous features in X_test using the same scaler
X_test_scaled = X_test.copy()  # Copy to keep original intact
X_test_scaled[continous_features] = scaler.transform(X_test[continous_features])
```
