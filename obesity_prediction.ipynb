<div style="text-align: center; border-top: 3px solid black; padding: 10px;">
    <h2><strong>4.Data Cleaning</strong></h2>
</div>

After using the info() and describe() functions for initial data inspection, we saw that there were no missing values. However, we will verify this further by explicitly checking for any missing values. Additionally, we will check for any duplicate rows in the dataset to ensure the data is clean.

```python
# verify if theres no missing rows

no_missing_rows = df.isna().sum()

print(no_missing_rows)
```

```python
# Number of duplicate rows
duplicate_count_before = df.duplicated().sum()

# print duplicate count
print('Duplicate row count :',duplicate_count_before)

# remove duplicate rows
df = df.drop_duplicates()

# Number of duplicate rows after deleting
duplicate_count_after = df.duplicated().sum()

# print duplicate count
print('Duplicate row count after dropping duplicates:',duplicate_count_after)
```

    Besides the dataset having duplicates, no missing values were found. After removing the duplicates, the dataset is now clean and free of any missing data and it is now ready for further analysis

#

<div style="text-align: center; border-top: 3px solid black; padding: 10px;">
    <h2><strong>4.Exploratoray Data Analysis</strong></h2>
</div>

<div style="text-align: center; padding: 10px;">
    <h3><strong>4.1 Univariate Analysis</strong></h3>
</div>

```python
# Univariate analysis - Categorical features (display all categorical features with 5 per row)
categorical_columns = ['Gender', 'family_history', 'FAVC', 'CAEC', 'MTRANS', 'Obesity']

# Set up subplots: 1 row, 6 columns
fig, axes = plt.subplots(1, 6, figsize=(18, 5))  # Wide figure to fit all plots

for i, col in enumerate(categorical_columns):
    sns.countplot(x=df[col], ax=axes[i])
    axes[i].set_title(f'{col}')
    axes[i].tick_params(axis='x', rotation=90)

plt.tight_layout(pad=1.5)  # Adjust padding for better spacing
plt.show()
```

- __Gender:__ The dataset seems to have a balanced distribution between Male and Female respondents.

- __Family History:__ A significant majority of individuals have a family history of obesity, which could be an important predictive factor.

- __Frequent Consumption of High-Calorie Food (FAVC):__ Most individuals in the dataset consume high-calorie food frequently, which aligns with obesity risk factors.

- __Consumption of Food Between Meals (CAEC):__ The majority of people in the dataset consume food "Sometimes," with smaller groups reporting frequent or always snacking.

- __Mode of Transportation (MTRANS):__ Most individuals use public transportation or walk, with fewer using motorbikes, automobiles, or bicycles.

- __Obesity:__ The distribution of obesity levels appears to be relatively balanced, though some categories have slightly higher counts,this is good as this featur will be later used as our target feature

```python
# Univariate analysis - Numerical features (display all numerical features with 4 per row)
numerical_columns = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']
fig, axes = plt.subplots(2, 4, figsize=(16, 8))  # 2 rows, 4 columns

for i, col in enumerate(numerical_columns):
    ax = axes[i // 4, i % 4]  # Calculate row and column
    sns.histplot(df[col], kde=True, ax=ax)
    ax.set_title(f'Distribution of {col}')

plt.tight_layout(pad=1.0)  # Adjust padding between plots
plt.show()
```

-  __Age:__ Presents a right-skewed plot meaning most individuals are relatively young,most being less than 30 years
- __NCP (Number of Main Meals per Day):__ Strong peaks at 3 meals per day, likely the standard.
- __FAF (Physical Activity Frequency per Week):__ Right-skewed, meaning many individuals have low physical activity.
- __FCVC (Frequency of Vegetable Consumption):__ Suprisingly Many individuals have high vegetable consumption.
